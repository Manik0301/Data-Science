```python
import requests
from bs4 import BeautifulSoup

url = "https://webscraper.io/test-sites/e-commerce/static"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Extract all headings
for tag in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']:
    headings = soup.find_all(tag)
    for heading in headings:
        print(tag.upper(), ":", heading.text.strip())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://webscraper.io/test-sites/e-commerce/static"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

links = []

for a in soup.find_all('a'):
    link = a.get('href')
    if link:
        links.append(link)

# Save to CSV
df = pd.DataFrame(links, columns=["Hyperlinks"])
df.to_csv("hyperlinks.csv", index=False)

print("Links saved successfully!")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://books.toscrape.com"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

products = soup.find_all("article", class_="product_pod")

data = []

for product in products:
    name = product.h3.a["title"]
    price = product.find("p", class_="price_color").text

    data.append([name, price])

df = pd.DataFrame(data, columns=["Product Name", "Price"])
df.to_csv("products.csv", index=False)

print("Product data saved!")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://news.ycombinator.com/newest"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

titles = soup.find_all("span", class_="titleline")

news_data = []

for title in titles:
    headline = title.text
    link = title.a["href"]

    news_data.append([headline, link])

df = pd.DataFrame(news_data, columns=["Headline", "Link"])
print(df)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.linkedin.com/"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

images = soup.find_all("img")

for img in images:
    alt_text = img.get("alt")
    if alt_text:
        print("Alt Text:", alt_text)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.bbc.com/"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Meta description
description = soup.find("meta", attrs={"name": "description"})

# Meta keywords
keywords = soup.find("meta", attrs={"name": "keywords"})

if description:
    print("Description:", description.get("content"))

if keywords:
    print("Keywords:", keywords.get("content"))
```

```python
#Loading weather dataset
import pandas as pd
df = pd.read_csv("weather(1).csv")
print(df)
```

```python
#Displaying first five rows of dataset
df.head()
```

```python
#Identifying missing values in the dataset
df.isnull().sum()
```

```python
#Replacing missing values in Avg Temp with mean
df["Data.Temperature.Avg Temp"].fillna(
    df["Data.Temperature.Avg Temp"].mean(),
    inplace=True
)
df.head()
```

```python
# replacing missing wind speed with average values
df["Data.Wind.Speed"].fillna(
    df["Data.Wind.Speed"].mean(),
    inplace=True
)
```

```python
#Verifying missing values are filled.
df.isnull().sum()
```

```python
# Convert Celsius to Fahrenheit
df["Data.Temperature.Avg Temp"] = (df["Data.Temperature.Avg Temp"] * 9/5) + 32
df['Data.Temperature.Avg Temp'].head()
```

```python
#converting and storing temperature in a new cloumn
df["Temperature_F"] = (df["Data.Temperature.Avg Temp"] * 9/5) + 32
df.head()
```

```python
#Min Max Normalization
df["Precipitation_MinMax"] = (
    (df["Data.Precipitation"] - df["Data.Precipitation"].min()) /
    (df["Data.Precipitation"].max() - df["Data.Precipitation"].min())
)
df['Precipitation_MinMax'].head()
```

```python
#Z Score Normalization
df["WindSpeed_Zscore"] = (
    (df["Data.Wind.Speed"] - df["Data.Wind.Speed"].mean()) /
    df["Data.Wind.Speed"].std()
)
df['WindSpeed_Zscore'].head()
```

```python
#Displaying Normalized Columns
df[["Precipitation_MinMax", "WindSpeed_Zscore"]].head()
```

```python
#Identifying Duplicate rows in dataset
print(df['Station.Location'].duplicated().sum())
```

```python
#Dropping duplicate rows
df = df.drop_duplicates()
df.shape
```

```python
#Displaying Total number of rows after removal
len(df)
```

```python
#Summary Statistics of cleaned dataset[]
df.describe()
```

```python
#Saving Cleaned Dataset
df.to_csv("cleaned_weather.csv", index=False)
```
